```python
from nltk.corpus import wordnet

print(wordnet.get_version())
```

```python
import pandas as pd

mturk771 = pd.read_csv("data/mturk-771.csv")[["word1", "word2", "similarity"]]
mturk771
```

```python
import gensim
model = gensim.models.KeyedVectors.load_word2vec_format("~/gensim-data/GoogleNews-vectors-negative300.bin.gz", binary=True)
```

```python
import pandas as pd
from tqdm import tqdm

def get_w2v_similarity_on_mturk771():
    data = mturk771.to_dict(orient="records")
    for item in tqdm(data):
        item["w2v"] = model.similarity(item["word1"], item["word2"])
    return data

result = pd.DataFrame(get_w2v_similarity_on_mturk771())
result
```

```python
result.to_csv("result/w2v_mturk771.csv")
```
